# NHANES 2021-2023 Analysis

## Loading in Files

```python
# Load in Files – Demographics

import pandas as pd

demographic_path = 'DEMO_L.XPT'
demo = pd.read_sas(demographic_path, format='xport') #, encoding='latin1')  # Try 'latin1' or 'iso-8859-1' if UTF-8 fails
demo
```

```python
# Load in Files – BP

bp_path = 'BPXO_L.XPT'
bp = pd.read_sas(bp_path, format='xport')
bp
```

```python
# Load in Files – Vitamin D Lab

vitamind_path = 'VID_L.XPT'
vitd = pd.read_sas(vitamind_path, format='xport')
vitd
```

```python
# Load in Files – Hepatitis B Lab

hepatitisb_path = 'HEPB_S_L.XPT'
hepb = pd.read_sas(hepatitisb_path, format='xport')
hepb
```

```python
# Load in Files – Weak/Failing Kidneys

kidneys_path = 'KIQ_U_L.XPT'
kidney = pd.read_sas(kidneys_path, format='xport')
kidney
```

```python
# Load in Files – Minutes of Sedentary Behavior

sedbehavior_path = 'PAQ_L.XPT'
behavior = pd.read_sas(sedbehavior_path, format='xport')
behavior
```

```python
# Load in Files – Current Self-Reported Weight

weight_path = 'WHQ_L.XPT'
weight = pd.read_sas(weight_path, format='xport')
weight
```

## Recoding variables

```python
# Marital Status, DMDMARTZ
demo['DMDMARTZ_binary'] = demo['DMDMARTZ'].apply(lambda x: 'Married' if x == 1 else 'Not Married')
demo['DMDMARTZ_binary']
```

```python
# Education Level, DMDEDUC2
demo['DMDEDUC2_binary'] = demo['DMDEDUC2'].apply(lambda x: 'Bachelors or higher' if x == 5 else 'Less than bachelors')
demo['DMDEDUC2_binary']
```

## Data cleaning

```python
# Hepatitis B Lab Antibodies, LBXHBS
## Two levels
print(hepb['LBXHBS'].unique())
hepb['LBXHBS_binary'] = hepb['LBXHBS'].apply(lambda x: 'Positive' if x == 1.0 else 'Negative')
hepb ['LBXHBS'].dropna()
```

```python
# Weak/Failing Kidneys, KIQ022
## Two levels
print(kidney['KIQ022'].unique())
kidney['KIQ022_binary'] = kidney['KIQ022'].apply(lambda x: 'Yes' if x == 1.0 else 'No')
kidney ['KIQ022'].dropna()
```

```python
# Minutes of Sedentary Behavior, PAD680
## Removing 7777,9999, nan values
behavior = behavior[~behavior['PAD680'].isin([7777, 9999])]
behavior = behavior.dropna(subset=['PAD680'])
behavior['PAD680'].tail(10)
```

```python
# Current Self-Reported Weight, WHD020
## Removing 7777,9999, nan values
weight = weight[~weight['WHD020'].isin([7777, 9999])]
weight = weight.dropna(subset=['WHD020'])
weight['WHD020'].tail(10)
```

## Questions for Analysis

```python
# Question 1: "Is there an association between marital status (married or not married) and education level (bachelor’s degree or higher vs. less than a bachelor’s degree)?"

import pandas as pd

contingency_table = pd.crosstab(demo['DMDMARTZ_binary'], demo['DMDEDUC2_binary'])
print(contingency_table)

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpretation
if p < 0.05:
    print("Significant association between marital status and education level.")
else:
    print("No significant association between marital status and education level.")
```

```python
# Question 2: "Is there a difference in the mean sedentary behavior time between those who are married and those who are not married?"

merged_data = pd.merge(demo[['SEQN', 'DMDMARTZ_binary']], behavior[['SEQN', 'PAD680']], on='SEQN')
from scipy.stats import ttest_ind

married_sedentary = merged_data[merged_data['DMDMARTZ_binary'] == 'Married']['PAD680']
not_married_sedentary = merged_data[merged_data['DMDMARTZ_binary'] == 'Not Married']['PAD680']

t_statistic, p_value = ttest_ind(married_sedentary, not_married_sedentary)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Significant difference in sedentary behavior time between married and not married individuals.")
else:
    print("No significant difference in sedentary behavior time between married and not married individuals.")
```

```python
# Question 3: "How do age and marital status affect systolic blood pressure?"

import statsmodels.formula.api as sm
!pip install statsmodels

merged_data = pd.merge(demo[['SEQN', 'RIDAGEYR', 'DMDMARTZ_binary']], bp[['SEQN', 'BPXOSY3']], on='SEQN')
model = sm.ols('BPXOSY3 ~ RIDAGEYR + DMDMARTZ_binary + RIDAGEYR * DMDMARTZ_binary', data=merged_data).fit()
print(model.summary())

# Interpretation
if model.pvalues['RIDAGEYR'] < 0.05:
    print("Significant effect of age on systolic blood pressure.")
else:
    print("No significant effect of age on systolic blood pressure.")
```

```python
# Question 4: "Is there a correlation between self-reported weight and minutes of sedentary behavior?"

merged_data = pd.merge(weight[['SEQN', 'WHD020']], behavior[['SEQN', 'PAD680']], on='SEQN')

from scipy.stats import pearsonr

correlation, p_value = pearsonr(merged_data['WHD020'], merged_data['PAD680'])

print(f"Pearson correlation coefficient: {correlation}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
  print("Significant linear relationship between self-reported weight and minutes of sedentary behavior.")
else:
  print("No significant linear relationship between self-reported weight and minutes of sedentary behavior.")
```

```python
# Question 5: "Is there a significant association between vitamin D deficiency and the prevalence of hepatitis B infection?"

merged_data = pd.merge(vitd[['SEQN', 'vitamin_d_deficient']], hepb[['SEQN', 'LBXHBS_binary']], on='SEQN')

contingency_table = pd.crosstab(merged_data['vitamin_d_deficient'], merged_data['LBXHBS_binary'])
print(contingency_table)

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpretation
if p < 0.05:
    print("Significant association between vitamin D deficiency and hepatitis B infection.")
else:
    print("No significant association between vitamin D deficiency and hepatitis B infection.")
```
